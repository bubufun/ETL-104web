```python
import pandas as pd
import jieba
import glob
import re
```

```python
txt_file_path =glob.glob(r'./104_web_ETL_collection/*.txt')
txt_file_path
```

```python
jieba.load_userdict('./HW-104-ETL-jieba-mydict')
```

```python
df3 = pd.read_csv(r'./HW-104-ETL-output.csv')
df3
```

```python
columnList = ['python',
              'java',
              'javascript',
              'r 語言',
              'mysql',
              'sql',
              'mongodb',
              'nosql',
              'aws',
              'gcp',
              'azure',
              'ai',
              'data mining',
              'deep learning',
              'cloud service']
for column in columnList:
    df3[column] = '0'
df3
```

```python
rowData_unanalysed = []
for article in txt_file_path:
    with open(article,'r',encoding='utf-8') as f:
        s = f.read().replace('\n',' ',).replace('---split---',' ')
        fil = re.compile(u'[^0-9a-zA-Z\u4e00-\u9fa5.，,。？“”]+', re.UNICODE)
        data = fil.sub(' ', s.upper())                                        #去掉非英文及中文字符
        new_data = ''.join(x for x in data if ord(x) < 256) #去掉中文
        rowData_unanalysed.append(new_data)
        
rowData_unanalysed
```

```python
rowData_analysed = []
for s in rowData_unanalysed:
    tmplist = jieba.cut(s)
    dataList = [i for i in tmplist]
    rowData_analysed.append(dataList)
rowData_analysed
```

```python
for index, article_content in enumerate(rowData_analysed):
    if 'PYTHON' in article_content:
        df3['python'][index] = '1'
    if 'JAVA' in article_content:
        df3['java'][index] = '1'
    if 'JAVASCRIPT' in article_content:
        df3['javascript'][index] ='1'
    if 'R LANGUAGE' in article_content:
        df3['r 語言'][index] ='1'
    if 'MYSQL' in article_content:
        df3['mysql'][index] ='1'
    if 'SQL' in article_content:
        df3['sql'][index] = '1'
    if 'MONGODB' in article_content:
        df3['mongodb'][index] ='1'
    if 'NOSQL' in article_content:
        df3['nosql'][index] ='1'
    if 'AWS' in article_content:
        df3['aws'][index] ='1'
    if 'GCP' in article_content:
        df3['gcp'][index] ='1'
    if 'AZURE' in article_content:
        df3['azure'][index] = '1'
    if 'AI' in article_content:
        df3['ai'][index] ='1'
    if 'DATA MINING' in article_content:
        df3['data mining'][index] = '1'
    if 'DEEP LEARNING' in article_content:
        df3['deep learning'][index] ='1'
    if 'CLOUD SERVICE' in article_content:
        df3['cloud service'][index] = '1'
df3
```

```python
df3.to_csv(r'./HW-104-ETL-data_analysis.csv',index =0,encoding='utf-8-sig')
```
