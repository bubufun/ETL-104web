```python
import pandas as pd
import glob
txt_file_path =glob.glob(r'./104_web_ETL_collection/*.txt')
txt_file_path
```

```python
rowData = []
for article in txt_file_path:
    with open(article,'r',encoding='utf-8') as f:
        tmplist = f.read().lstrip('---split---').split('---split---')
#         newList =[i for i in enumerate(tmplist)]
        rowData.append(tmplist)
rowData
```

```python
columns = ['title','company','titleUrl','job_content','job_require','job_welfare','job_contact']
columns
```

```python
df = pd.DataFrame(columns=columns,data=rowData)
df
```

```python
df2 =df.copy()
for c in df2:
    df2[c] = df2[c].apply(lambda x : x.replace('\n',''))
df2
```

```python
df2.to_csv(r'./HW-104-ETL-output.csv',index =0,encoding='utf-8-sig')
```
